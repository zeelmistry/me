This dataset has 9564 rows and 49 columns

This dataset provides information on ‘Kepler Objects of Interest’ which are exoplanets that are being observed by the Kepler Telescope that was launched into space. This information divides based on which stellar (solar) system, then to each planet in that subdivision. The information consists of candidates categorised if they are similar and habitable like earth with other information such as surface temperature, gravitational strength, disposition and so on. 


This data set was collected for the overall reason to search for similar earth-like planets in their respective systems away from ours. The research and data’s objective was to:


o	Determine the amount of exo-planets that are near or in the habitable zone of other stellar systems


o	Determine the size, distribution and shapes of these orbits


o	Determine amount of other planets in that stellar system


o	Determine the properties of the parent stars in that system


This data was collected by the Kepler Space Telescope which was funded and launched by NASA and was collected from NASA and its website.

```python
def bar_graph(datachoice, xvalue, widthchoice, colorchoice, labelchoice):
    plt.bar(xvalue, datachoice, width = widthchoice, color = colorchoice, label = labelchoice)

def multiple_bar_graph(numberofbars, datachoices, xvalues, widths, colorchoices, labelchoices):
    for i in numberofbars:
        bar_graph(datachoices[i],xvalues[i], widths, colorchoices[i], labelchoices[i])
```

```python
def histogram_graph(datachoice, binvalue, colorchoice):
    plt.hist(datachoice, bins = binvalue, facecolor = colorchoice)

def histogram_range_graph(datachoice, binvalue, colorchoice, low,high):
    plt.hist(datachoice, bins = binvalue, facecolor = colorchoice, range = (low, high))

def histogram_comparison_graph(datachoice, binvalue1, binvalue2, colorchoice1, colorchoice2, low, high):
    histogram_graph(datachoice, binvalue1, colorchoice1)
    histogram_range_graph(datachoice, binvalue2, colorchoice2, low, high)
```

```python
def scatter_graph(datachoice, xvalue, sizechoice, colorchoice):
    plt.scatter(xvalue, datachoice, s = sizechoice, c = colorchoice)
```

```python
def pie_graph(dataname, colorchoice, explosion):
    dataname.plot(kind = 'pie', autopct = '%1.1f%%', colors = colorchoice, explode = explosion)
```

```python
def line_graph(datachoice, colorchoice):
    datachoice.plot.line(color = colorchoice)
```

```python
def graph_labels(xlabel, ylabel, title, trueorfalse):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(trueorfalse)
```

### Filtering process for these habitable planets

```python
kepler_data["kepoi_name"].value_counts()
```

There are so far within the data, 9564 potential habitable planetary objects transitting various stars, these now need to be filtered by the false positive flags within the transit process. 

### False Positive Flags

These false positive flags range from different issues through technical or natural phenomena. The filters in this data include; 

Not Transit-Like Flag: measures the light curve and sees whether that light curve is not like a transiting planet and another light source is warping the collected information

Stellar Eclipse Flag: observes if the transit has a significant secondary event, transit shape, or out-of-eclipse variability which suggests there is a discrepancy due to two stars (binary) or hot gas giants. 


Centroid Offset False Positive Flag: observes whether measuring the centroid was offset causing disruptions in collected information

Ephemeris Match Indicates Contamination False Positive Flag: observes whether the specific object of interest has the same transit as another object which ends up covering another potential object, this can be determined through the flux contamination

```python
#Data
ec = kepler_data["koi_fpflag_ec"].value_counts()
co = kepler_data["koi_fpflag_co"].value_counts()
ss = kepler_data["koi_fpflag_ss"].value_counts()
nt = kepler_data["koi_fpflag_nt"].value_counts()
#Variables for Function
xrange = np.arange(2)
xspread = [xrange-0.075, xrange-0.025, xrange+0.025,xrange+0.075]
bars = [0,1,2,3]
dataspread = [ss, co, ec, nt]
widthspread = 0.05
colorspread = ['#330147','#54006b','#6b0060','#7d0055']
labelspread = ['Stellar Eclipse', 'Centroid Offset', 'Ephemeris Match Indicates Contamination','Not Transit-Like Flag']
#Bar Function
multiple_bar_graph(bars, dataspread, xspread, widthspread, colorspread, labelspread)
#Labels and Tags
plt.legend()
graph_labels("Types","Number of False Positive Flags",
             "False Positive Flags", False)
```

As it can be seen, most false positive flag are 0 suggesting that the flags are not being triggered. This spans across all four false positive flags with the Stellar Eclipse Flag being registered the most and Ephemeris Contamination Flag with the least registered. This is data does reflect natural phenomena as the chances of two perfectly transiting objects is furthermore rare than binary stars or hot gas giants in a stellar system causing differences in data collection.

### The Filter Process

```python
stellarandearthlike_filtered = confirmed_planet[(confirmed_planet.koi_period < 400) & (confirmed_planet.koi_period > 300) & (confirmed_planet.koi_prad < 1.1) & (confirmed_planet.koi_prad > 0.9) & (confirmed_planet.koi_teq < 300) &  (confirmed_planet.koi_teq >200) & (confirmed_planet.koi_steff < 6400) & (confirmed_planet.koi_steff >5200 ) & (confirmed_planet.koi_slogg < 5) & (confirmed_planet.koi_slogg >4 ) & (confirmed_planet.koi_srad < 1.3) & (confirmed_planet.koi_srad >0.7 )]
print(stellarandearthlike_filtered)
```

As it can be seen, with these specific Earth-like restrictions, there seems to be only one other exoplanet that has been discovered by the Kepler Telescope that is the most similar to Earth. With it having similar G-Type Star with similar size and gravity with the planet also having a close orbital period and radius. 

Obviously there are many other Earth-like exoplanets, some called Super Earths and such. These include Kepler 442 - b and Kepler 22 - b which are popularly known however they orbit different star types such as K type and such which are different to G Star type Sun. 

Kepler - 452 b however, is an exoplanet that could be a potential rocky Earth that orbits a star extremely similar to the Sun. It lies almost 1400 light years away and has very similar features. 

```python
#Data
kepler452b_period = stellarandearthlike_filtered["koi_period"]
kepler452b_prad = stellarandearthlike_filtered["koi_prad"]
kepler452b_teq = stellarandearthlike_filtered["koi_teq"]
earth_period = earth_data["koi_period"]
earth_prad = earth_data["koi_prad"]
earth_teq = earth_data["koi_teq"]
#Variables for Function
xspread = [0-0.25, 0+0.25, 2-0.25,2+0.25,4-0.25,4+0.25]
bars = [0,1,2,3,4,5]
dataspread = [kepler452b_period, earth_period, kepler452b_teq, earth_teq, kepler452b_prad, earth_prad]
widthspread = 0.5
colorspread = ['#3f2642','#4a1c4f','#080242','#371e5e','#75085f','#75083d']
labelspread = ['a','b','c','d','e','f']
#Bar Function
multiple_bar_graph(bars, dataspread, xspread, widthspread, colorspread, labelspread)
#Graph Labels
graph_labels("","value","Comparison of Earth and Kepler 452 b", False)
#Values and Tags
basedistance = [0-0.295,0+0.225, 2-0.295,2+0.225,4-0.295,4+0.225]
basevalue = [385, 365, 220, 255, 1.09, 1]
baselabel = ["Kepler 452 b Orbital Period", "Earth Orbital Period", "Kepler 452 b Temperature",
            "Earth Temperature", "Kepler 452 b Radius", "Earth Radius"]
for i in range(0,6):
    plt.text(basedistance[i], basevalue[i]+5, s = basevalue[i])
    plt.text(basedistance[i], -15, s = baselabel[i], rotation = 90)
```

The temperature is lower due to its material make up however the gravity and such would be extremely similar to that of Earth. However, it seems that the surface is quite active with volcanoes and a much thicker atmosphere due to its density and mass.
